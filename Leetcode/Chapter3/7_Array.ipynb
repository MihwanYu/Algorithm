##배열
배열은 value 또는 variable element의 집합으로 구성된 구조, 하나 이상의 index 또는 key로 식별된다.

자료구조
1. 메모리 공간 기반의 연속(contiguous)방식
2. 포인터 기반 연결(link)방식 ex)연결리스트

추상자료형(Abstract Data Type): 배열 or연결리스트 기반

동적배열:

크기를 지정하지 않도 자동으로 리사이징하는 배열

자바 -ArrayList, C++ -std::vector, 파이썬 -list

원리: 미리 초깃값을 작게 잡아 배열 생성, 데이터 추가하면서 채워지면, 배열 크기를 늘려주고 모두 복사(Doubling)
