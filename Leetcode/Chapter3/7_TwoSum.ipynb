```python
#덧셈하여 타겟을 만들 수 있는 배열의 두 숫자 인덱스를 리턴하라
nums = [2,7,11,15]
target=9
```

```python
for i in range(len(nums)):
  for j in range(i+1, len(nums)):
    if nums[i]+nums[j] ==target:
      answer = (i,j)

print(answer)

#교재풀이1과 같음. 시간복잡도가 O(n^2)-브루트 포스 방식
#모든 조합을 비교하지 않음
```

```python
#교재풀이2) in을 이용한 탐색
#시간복잡도는 O(n^2)지만 in연산이 빠르기 때문에 1번보다 더 빠른 방법임

def twoSum(nums,target):
  for i, n in enumerate(nums):
    #complement에 9를 만들 수 있는 다른 숫자 짝을 저장, 그 숫자가 리스트의 나머지 요소 중에 있는지 확인
    complement = target-n

    if complement in nums[i+1:]:
      return nums.index(n), nums[i+1:].index(complement)+(i+1)
```

```python
twoSum(nums, target)
```

```python
#교재풀이3) 첫 번째 수를 뺀 결과 키 조회
#딕셔너리는 해시 테이블로 구현-> 조회시간이 평균 O(1), 최악이 O(n)임
def twoSum(num, target):
  nums_map={}

  #키와 값 바꿔서 딕셔너리로 저장
  for i, num in enumerate(nums):
    nums_map[num] = i
  
  #타겟에서 첫 번째 수를 뺀 결과를 키로 조회
  #그러면 두번째 수를 바로 알아낼 수 있으니까! 두번째 수를 키로 조회해서 인덱스 번호 찾기
  for i,num in enumerate(nums):
    if target-num in nums_map and i!= nums_map[target-num]:
      return nums.index(num), nums_map[target-num]
```

```python
#교재풀이4) 조회 구조 개선
#3번의 딕셔너리 저장, 조회를 for문 1개로 합쳐서 처리해보기(성능상 차이 크지x)

def twoSum(num, target):
  nums_map={}

  for i, num in enumerate(nums):
    if target-num in nums_map:
      return [nums_map[target-num], i]
    nums_map[num] = i
```

```python
#교재풀이5) 투 포인터 이용
#투 포인터: 왼쪽 포인터,오른쪽 포인터 합이 타겟보다 크면 오른쪽 포인터를 왼쪽으로, 작다면 왼쪽 포인터를 오른쪽으로 옮기며 값을 조정하는 방식
#근데 이거로 풀려면 nums가 정렬이 된 상태여야함. 뒤에서 다시 보는거로~(p.189 박스)
```
