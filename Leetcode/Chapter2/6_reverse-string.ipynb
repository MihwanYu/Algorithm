알고리즘 인터뷰 6-1장 <문자열 뒤집기>

```python
#문자열 뒤집는 함수: 입력값은 문자 배열, 리턴 없이 리스트 내부를 직접 조작
#[h,e,l,l,o]->[o,l,l,e,h]

slist = ['h','e','l','l','o']
# length = len(slist)
i = 0
while i <= int(len(slist)/2):
  temp = slist[i]
  slist[i] = slist[len(slist)-1-i]
  slist[len(slist)-1-i] = temp
  i += 1
print(slist)
```

```python
def reverse_str(slist: str)-> list:
  i = 0
  while i <= int(len(slist)/2):
    temp = slist[i]
    slist[i] = slist[len(slist)-1-i]
    slist[len(slist)-1-i] = temp
    i += 1

  return slist
```

```python
case1 = ['h','e','l','l','o']
case2 = ['H','a','n','n','a','h']
print(reverse_str(case1))
print(reverse_str(case2))
```

```python
#교재 풀이 1) 투 포인터 이용한 swap
def reverseString(s)->None:
  left, right = 0, len(s)-1 #앟...... 이렇게 한번에 두 변수 선언이 되지 참,,,
  while left<right:
    s[left],s[right] = s[right],s[left] #이렇게 표현하는 방법도 있구만...
    left+=1
    right-=1
```

```python
case1 = ['h','e','l','l','o']
case2 = ['H','a','n','n','a','h']
reverseString(case1)
reverseString(case2)
print(case1)
print(case2)
```
