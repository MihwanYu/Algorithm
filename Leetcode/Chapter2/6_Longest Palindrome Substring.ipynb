```python
#babad->bab or aba
#cbbd->bb
```

```python
strs='babad'


samechar=[]# 같은 문자를 가진 애들끼리 문자열 인덱스 값 저장해놓기
i = 1
for i in range(len(strs)):
  print('지금 s는 ',strs[i])
  
  for j in range(len(strs[i+1:])):
      # print(j)
      # print(strs[i+1:][j])
      if strs[i]==strs[i+1:][j]:
        print(strs[i+1:][j])
        samechar.append((i,j+i+1))
      # strs=strs[i:]
      # print('지금 strs는 ',strs)

samechar
```

```python

palindrome={}#samechar 기준으로 가능한 모든 palindrome 저장
for i,j in samechar:
    palin=''
    k=(i+j)//2
    # print(k)
    if (j-i)%2==0: #사잇값이 홀수일때
        palin+=strs[k]
        front=back=k
    else: #사잇값이 짝수일때 abba
        palin += strs[k+1]+strs[k]
        front=k-1
        back=k
    while front>0 and back<j:
        front -=1
        back +=1
        if strs[front]!=strs[back]:
            break
        palin=strs[front]+palin+strs[back]

    print(palin)
    palindrome[palin]=len(palin)

longest_palin = max(palindrome, key=palindrome.get) #딕셔너리 key 사용해서 값 추출하기!!!익숙해질것!!!
print('가장 긴 팰린드롬: ',longest_palin)
```

처음부터 끝까지, 뒤로 한칸씩--> 문자열 순서 거꾸로

```python
strs='babad'


strs[::-1]
```

```python
#교재해설) 2개의포인터
def longestPalindrome(s):
  
  def expand(left:int, right: int):
    while left>=0 and right<=len(s) and s[left]==s[right-1]:
      left-=1
      right+=1
    return s[left+1:right-1]

  #문자열이 1개짜리거나, 전체 문자열을 거꾸로 해도 값이 같을때 곧바로 return s
  if len(s)<2 or s==s[::-1]:
    return s

  #그게 아니면 이제 빈 문자열 만들어서 채워나가는거지..
  result=''
  for i in range(len(s)-1):
    #expand(i,i+1)은 palindrome 길이 짝수일경우, expand(i,i+2)는 홀수일경우를 리턴
    #max()함수 안에 우겨넣음으로써 앞에 있던 result와 길이 비교하고 가장 길이 긴 문자열 리턴
    result=max(result,expand(i,i+1),expand(i,i+2),key=len)
  return result
```

```python
longestPalindrome(strs)
```
