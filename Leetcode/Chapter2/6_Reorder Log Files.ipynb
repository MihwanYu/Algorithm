알고리즘 인터뷰 6-1 <로그 파일 재정렬>

```python
#가장 앞 부분은 식별자
#문자로 구성된 로그가 숫자보다 앞
#식별자는 순서에 영향x, 문자가 동일하면 식별자 순
#숫자는 입력순
#이게뭔소리야,...
```

```python
logs=["dig1 8 1 5 1","let1 art can","dig2 3 6","let2 own kit dig","let3 art zero"]
#dig1, let1 쟤네가 식별자

def reorderLogFiles(logs):
  letters,digits=[],[]
  for log in logs:
    if log.split()[1].isdigit():
      digits.append(log)
    else:
      letters.append(log)

  letters.sort(key=lambda x:(x.split()[1:],x.split()[0]))
  #식별자를 제외한 문자열 [1:]를 키로 하여 정렬, 동일한 경우 후순위로 식별자 [0]을 지정해 정렬
  return letters+digits
```

```python
result = reorderLogFiles(logs)
result
```

sort()의 매개변수 key

정렬을 목적으로 하는 함수를 값으로 넣는다. lambda를 이용할 수 있다.

key 값을 기준으로 정렬되고 기본값은 오름차순이다. 

여러 개의 요소를 가진 경우, 튜플로 사용할 수 있다.
