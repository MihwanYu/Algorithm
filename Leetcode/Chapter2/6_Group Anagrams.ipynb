````python
i = 0
while i<len(origin):
    
    group = []
    group.append(origin[i])
    for word in origin[i+1:]:
      if sorted(list(origin[i]))==sorted(list(word)):
        # print(word)
        group.append(word)
        origin.remove(word)
        
    # print(group)
    anagram.append(group)
    i+=1


print(anagram)
````

##파이썬의 정렬 방법
sorted: 문자, 리스트의 정렬

```
a=[2,5,1,9,7]
sorted(a)
```

sorted vs sort

1. sorted(): 정렬 결과를 별도로 리턴함-다른 변수에 리턴결과 저장가능
2. sort(): 리스트 자체를 제자리 정렬. None을 리턴한다

sorted의 key 파라미터:

정렬을 위한 키 또는 함수를 별도로 지정한다.
```
c=['ccc','aaaa','d','bb']
sorted(c, key=len)
```
이 경우 길이를 구하는 len 함수가 key가 되어 길이 순서로 정렬됨
```
a=['cde','cfc','abc']
def fn(s): #fn()은 첫 문자열과 마지막 문자열을 리턴함
  return s[0], s[-1]
print(sorted(a,key=fn))
```
이 경우에는 첫 문자열과 마지막 문자열 순으로 정렬이 된다


똑같은 정렬을 람다식 사용하기:
```
a=['cde','cfc','abc']
sorted(a, key=lambda s:(s[0], s[-1]))
```
