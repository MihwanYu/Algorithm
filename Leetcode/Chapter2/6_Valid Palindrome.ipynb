```python
def Palindrome(submit):
  submit = submit.lower() #change to lowercase
  str_li = [c for c in submit if c.isalnum()==True] #make list only having alpha, num
  length = len(str_li)
  for i in range(int(length/2)):
    if (str_li[i] != str_li[length-1-i]):
      return False
  return True
```

```python
#교재 풀이 1) 리스트로 변환, pop과 while 사용
def isPalindrome(s: str)-> bool: #여기 보면 s가 str 타입이라고 명시해놨음 + 반환형도 아예 지정해놓은 모습
  strs = []
  for char in s:
    if char.isalnum():
      strs.append(char.lower())
  while len(strs)>1:
    if(strs.pop(0)!=strs.pop()): #list 에서는 pop에다가 인덱스 지정 가능함!-> O(n)
      return False
  return True
```

```python
subs = 'A man, a plan, a canal: Panama'
# subs = 'race a car'
answer = isPalindrome(subs)
print(answer)
```

```python
#교재 풀이 2) 데크 자료형을 이용한 최적화
def isPalindrome2(s:str)-> bool:
  strs: Deque = collections.deque() #자료형을 Deque로 선언
  for char in s:
    if char.isalnum():
      strs.append(char.lower())
  while len(strs)>1:
    if(strs.popleft!=strs.pop()): #Deque의 popleft를 사용함!-> O(1)
      return False
  return True
```
