알고리즘 인터뷰 6-1 <가장 흔한 단어>

```python
#금지된 단어를 제외한 가장 흔하게 등장하는 단어 출력
#대소문자 구분x
#구두점(마침표,쉼표) 무시
```

```python
paragraph = "Bob hit a ball, the hit BALL flew far after it was hit."
banned = ["hit"]
```

```python
from collections import Counter
import re

# para_li = paragraph.lower()
para_li = re.sub(r'[^\w]', ' ', paragraph).lower().split()
para_li = [word for word in para_li if word not in banned]
word_dic = Counter(para_li)
# print(word_dic)
max(word_dic, key=word_dic.get) #max 함수에도 매개변수 key를 사용할수있다..
word_dic.most_common(1) #근데 애초에 Counter 모듈에서는 이렇게 하는게 가능하지!!
```
